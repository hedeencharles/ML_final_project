# Import Dataset and describe

```python
# libraries
import pandas as pd 
import matplotlib as plt
import seaborn as sns
import numpy as np
```

```python
dataset = pd.read_excel("../Datasets/insurance.csv")

# drop patient ID column
dataset = dataset.drop(columns=['Patient Id'])

dataset.columns
```

```python
# Explore the features, identify any significances
```

```python
# run dataset through info() and describe() 
dataset.info()
# dataset.describe()
```

# Data Distribution

```python
# What is the count of each level?
sns.countplot(x='Level',data=dataset)
```

```python
# What does the age distribution look like for each level?
sns.boxplot(x="Level", y="Age", data=dataset,palette='rainbow')
```

```python
# What levels smoke more?
sns.violinplot(x="Level", y="Smoking", data=dataset,palette='rainbow')
# sns.boxplot(x="Level", y="Smoking", data=dataset,palette='rainbow')
```

```python
# What about passive smoking?
# What does the age distribution look like for each level?
sns.violinplot(x="Level", y="Passive Smoker", data=dataset,palette='rainbow')
# sns.boxplot(x="Level", y="Passive Smoker", data=dataset,palette='rainbow')
```

```python
# Can do entire dataframe with orient='h'
sns.boxplot(data=dataset,palette='rainbow',orient='h')
```

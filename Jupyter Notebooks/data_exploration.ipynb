# Data Distribution

```python
# What is the count of each age?
sns.countplot(x='age',data=dataset)
```

```python
# What is the count of each smokers(y/n?
sns.countplot(x='smoker',data=dataset)
```

```python
# What is the count of each children?
sns.countplot(x='children',data=dataset)
```

```python
# What is the count of each region?
sns.countplot(x='region',data=dataset)
```

```python
# What does the age distribution look like for each level?
sns.boxplot(x="charges", y="age", data=dataset,palette='rainbow')
```

```python
# Can do entire dataframe with orient='h'
dataset_dropped_charges = dataset.drop(columns=['charges'])
sns.boxplot(data=dataset_dropped_charges,palette='rainbow',orient='h')
```

```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('../../Datasets/insurance.csv')
dataset
```

```python
# X will be the independent variables (features containing info)
X = dataset.iloc[:, :-1].values
# y will be the dependent variable vector (Purchased?)
y = dataset.iloc[:, -1].values
```

```python
# Check which columns need to be encoded (columns containing categorical/string data)

# dataset.dtypes
## OR
categorical_columns = [ col for col, dt in dataset.dtypes.items() if dt == object]
categorical_columns

def column_index(df, query_cols):
    cols = df.columns.values
    sidx = np.argsort(cols)
    return sidx[np.searchsorted(cols,query_cols,sorter=sidx)]

columns_to_encode = column_index(dataset, categorical_columns)
print(columns_to_encode)
```

```python
# One hot encoding: turn string column into 3 different columns (countries), for 3 different categories
# Bindary vector: each country to a certain order of the columns

# Coding independent variable
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

# Create object of the column transformer class
# [0] is the index of column to apply OneHotEncoding
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [1,4,5])], remainder='passthrough')

# Apply the transform method to change the column X
X = np.array(ct.fit_transform(X))

print(X)
```
